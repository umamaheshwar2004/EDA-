```python
display(data['region-code'].value_counts())
```

```python
country=data["region-code"].value_counts().head(5).values
Year=data["region-code"].value_counts().head(5).index
plt.gcf().set_size_inches(20,11)
plt.pie(country,labels=Year,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'},explode=[0,0,0,0,0])
plt.title('COUNTRY CODES')
plt.show()
```
